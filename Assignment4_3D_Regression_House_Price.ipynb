# 3D Regression Plot for House Price Prediction


**Write a Python program to draw a 3D plot that visualizes the regression model for house price prediction using suitable Python-based 3D plotting libraries.**



**Assume the following features were used:**

**Area (sq ft)**

**Number of Bedrooms**

**House Price**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Sample dataset
area = np.array([800, 1000, 1200, 1500, 1800, 2000])
bedrooms = np.array([1, 2, 2, 3, 3, 4])
price = np.array([40, 50, 60, 75, 90, 110])  # in lakhs

# Reshape data
X = np.column_stack((area, bedrooms))
y = price

# Train regression model
model = LinearRegression()
model.fit(X, y)

# Create meshgrid for regression plane
area_range = np.linspace(area.min(), area.max(), 20)
bedroom_range = np.linspace(bedrooms.min(), bedrooms.max(), 20)
area_grid, bedroom_grid = np.meshgrid(area_range, bedroom_range)

# Predict prices for grid
price_pred = model.predict(
    np.column_stack((area_grid.ravel(), bedroom_grid.ravel()))
).reshape(area_grid.shape)

# Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot (actual data)
ax.scatter(area, bedrooms, price, marker='o')

# Regression surface
ax.plot_surface(area_grid, bedroom_grid, price_pred, alpha=0.6)

# Labels
ax.set_xlabel("Area (sq ft)")
ax.set_ylabel("Number of Bedrooms")
ax.set_zlabel("House Price")
ax.set_title("3D Visualization of House Price Regression Model")

plt.show()
```

Dots  Actual house price data

Surface  Regression model predicting prices

Shows how area and bedrooms together influence price

The 3D plot visualizes the regression model where house price is predicted using area and number of bedrooms. The scatter points represent actual data, while the surface represents the regression plane learned by the model.
