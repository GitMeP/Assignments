# Neural Network Architecture for Fraud Detection



**Write a Python program to draw (visualize) the architecture of a Neural Network used to classify fraudulent and non-fraudulent credit card transactions.**

**Assume the fraud detection dataset contains the following input features:**



**Transaction Amount**

**Transaction Time**

**Merchant Category**

**Customer Age**

**Account Balance**

**Number Of Transactions Today**

**Fraud (0 = Genuine, 1 = Fraud)**

```python
import matplotlib.pyplot as plt

def draw_neural_network(layer_sizes):
    """
    Draws a neural network architecture.
    layer_sizes: list containing number of neurons in each layer
    """

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.axis('off')

    x_spacing = 2
    y_spacing = 1

    for layer_idx, layer_size in enumerate(layer_sizes):
        x = layer_idx * x_spacing
        y_start = -(layer_size - 1) * y_spacing / 2

        # Draw neurons
        for neuron_idx in range(layer_size):
            y = y_start + neuron_idx * y_spacing
            circle = plt.Circle((x, y), 0.15, fill=True)
            ax.add_patch(circle)

            # Draw connections
            if layer_idx > 0:
                prev_layer_size = layer_sizes[layer_idx - 1]
                prev_y_start = -(prev_layer_size - 1) * y_spacing / 2
                prev_x = (layer_idx - 1) * x_spacing

                for prev_neuron_idx in range(prev_layer_size):
                    prev_y = prev_y_start + prev_neuron_idx * y_spacing
                    ax.plot([prev_x, x], [prev_y, y], linewidth=0.5)

    # Layer labels
    labels = [
        "Input Layer\n(6 Features)",
        "Hidden Layer 1\n(8 Neurons)",
        "Hidden Layer 2\n(4 Neurons)",
        "Output Layer\n(Fraud / Genuine)"
    ]

    for i, label in enumerate(labels):
        ax.text(i * x_spacing, -3.5, label, ha='center', fontsize=10)

    plt.title("Neural Network Architecture for Credit Card Fraud Detection", fontsize=14)
    plt.show()


# Define network structure
layer_sizes = [6, 8, 4, 1]

# Draw the neural network
draw_neural_network(layer_sizes)
```

The neural network consists of an input layer with six neurons representing transaction-related features, two hidden layers for learning complex patterns, and an output layer with one neuron for binary fraud classification. The architecture is visualized using matplotlib.
